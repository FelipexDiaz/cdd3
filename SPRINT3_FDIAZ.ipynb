```python
import pandas as pd


# aqui se crea el dateframe
datos = {'Calorias': [420, 380, 390, 499, 300],'Tiempo': [60, 40, 75, 55, 45]}
# aqui se indexa el dataframe
df = pd.DataFrame(datos, index= ['L', 'M', 'X', 'J', 'V'])
# aqui se muestra el dataframe
display (df)


```

2. Calcular la media, mediana, y desviación típica de ambas columnas.

```python

media = df.mean()
mediana = df.median()
desviacion = df.std()

# aqui se realizan los calculos
calculos = pd.concat([media, mediana, desviacion], axis = 1)
calculos.columns = ['Media', 'Mediana', 'Desviación']
print(calculos)
```

```python
import pandas as pd

# aqui se agrega la columna al dataframe anterior
df['Reto'] = (df['Calorias'] / df['Tiempo']) > 400 / 60

display(df)
```

image.png

```python
# aqui se crea otro dataframe con los datos filtrados
df_filtrado = df.iloc[::2][(df['Calorias'] > 400)]

display(df_filtrado)
```

```python


df2 = df.groupby(['Reto']).count()
df2.index = df2.index.map({True: 'Cumplidos', False: 'No Cumplidos'})

# aqui se crea la columna con los porcentajes
df2['%'] = 100 * df2['Calorias'] / df2['Calorias'].sum()

# aqui se asigna la columna a la serie
Serie_porcentajes=df2['%']

display(Serie_porcentajes)

```

```python
import matplotlib.pyplot as plt


# aqui se definen los datos

dias = df.index
calorias = df[ 'Calorias']
tiempo = df['Tiempo']


# aqui se define el tamanno del grafico
plt.figure(figsize=(10, 6))


# aqui se definen las curvas del grafico
plt.scatter(dias, calorias, color='red', label='Calorías')
plt.plot(dias, calorias, color='red', alpha=0.5)
plt.scatter(dias, tiempo, color='blue', label='Tiempo')
plt.plot(dias, tiempo, color='blue', alpha=0.5)

# aqui se definen los titulos del grafico
plt.title('Progresión de Calorías y Tiempo durante la Semana')
plt.xlabel('Días')
plt.ylabel('Cantidad')

# aqui se define la posicion de la leyenda
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```
